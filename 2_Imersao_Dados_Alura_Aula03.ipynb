## **Desafio 1: Plotar as médias, medianas e moda nas notas de LC e MT (matiplotlib linha vertical)**

```
# Funções para cálculo de média, mediana e moda no Pandas: mean(), median() e mode(), respectivamente
# Para plotar as linhas verticais, usamos a função plt.vlines(), passando a média/mediana/moda das
# notas como parãmetro. O parâmetro ymax foi definido analisando os valores do gráfico
#
# Tamanho da figura
#
plt.figure(figsize=(12,8))

# Plotando histograma do Seaborn com a distribuição das notas (excluindo zeros na nota total)
# da prova de matemática
#
sns.histplot(dados_sem_notas_zero, x = "NU_NOTA_MT")

# Plotando linha vertical com o valor da média(mean)
#
plt.vlines(dados_sem_notas_zero['NU_NOTA_MT'].mean(),ymin=0, ymax=1900, color='r', linestyles='dashed')

# Plotando linha vertical com o valor da mediana(median)
#
plt.vlines(dados_sem_notas_zero['NU_NOTA_MT'].median(),ymin=0, ymax=2160, color='y', linestyles='dashed')

# Plotando linha vertical com o valor da moda (mode)
#
plt.vlines(dados_sem_notas_zero['NU_NOTA_MT'].mode(),ymin=0, ymax=2780, color='tab:orange', linestyles='dashed')

# Adicionando título ao gráfico
#
plt.title('Histograma das notas da prova de Matemática: média (vermelho), mediana (amarelo) e moda (laranja)')
```

**Pela definição, temos uma distribuição assimétrica à direita para as notas de Matemática: média e mediana estão à direita da moda.**

##**Desafio 4: Pensar sobre a correlação entre Matemática e Linguagens**

**Pelos dados, percebe-se que há uma certa correlação entre as notas de Matemática e Linguagens, que geralmente são vistas como áreas contrárias: as pessoas com aptidão em Exatas tem maior facilidade em Matemática e Ciências da Natureza, enquanto as pessoas com maior aptidão em Linguagens e Códigos se saem melhor em Ciências Humanas. O que explicaria então esse fato?**

**1) Uma hipótese que podemos traçar é de que essa correlação parece ser mais linear à medida que a nota aumenta. Em outras palavras, a pessoa que se prepara mais e estuda mais para a prova, em geral, estuda todas as áreas e tende a se sair melhor em todas elas. A pessoa pode ser melhor em Matemática ou Linguagens, mas compensa a área mais fraca com mais empenho nos estudos e consegue fazer uma boa prova no geral.**
